```python
import numpy as np
import cv2
import matplotlib.pylab as plt
import sys
```

```python
print('Hello OpenCV', cv2.__version__)
```

```python
img=cv2.imread('/content/S_7.jpg') #이미지를 불러와서 img에 리턴한다.
```

```python
if img is None: #이미지가 만약 없으면
    print('Image load failed!')
    sys.exit() #종료한다.
```

```python
# cv2.namedWindow('image')
# cv2.imshow('image',img)
# cv2.waitKey()
# cv2.destroyAllWindows()

# Display the image using matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Convert BGR to RGB for matplotlib
plt.axis('off') # Hide axes
plt.show()
```

Colab 환경에서는 OpenCV GUI 함수(cv2.namedWindow, cv2.imshow, cv2.waitKey, cv2.destroyAllWindows)를 사용할 수 없음을 나타냅니다.

이 함수들은 별도의 창에 이미지를 표시하는 데 사용되지만, Google Colab과 같은 웹 기반 환경에서는 지원되지 않습니다.

Colab에서 이미지를 표시하려면 matplotlib.pyplot.imshow를 사용할 수 있습니다.
