```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Colab에서는 cv2.imshow 대신 사용
```

```python
# 예제 이미지를 다운로드합니다. (이미지가 없다면 직접 업로드해도 됩니다.)
import urllib.request

url = 'https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg'
urllib.request.urlretrieve(url, 'lenna.bmp')
```

## 그레이스케일 영상 불러오기 및 밝기 조절

```python
# lenna.bmp 이미지를 그레이스케일로 불러옵니다.
src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
else:
    # 이미지를 밝게 만들기: 모든 픽셀 값에 100을 더함
    # np.clip을 사용해 값이 0보다 작아지거나 255보다 커지는 것을 막음
    dst = np.clip(src + 100., 0, 255).astype(np.uint8)

    # 원본 이미지와 밝기 조절 이미지를 각각 출력
    print('원본(그레이스케일) 이미지')
    cv2_imshow(src)
    print('밝기 증가 후 이미지')
    cv2_imshow(dst)
```

## 컬러 영상 불러오기 및 밝기 조절

```python
# lenna.bmp 이미지를 컬러로 불러옵니다.
src_color = cv2.imread('lenna.bmp')

if src_color is None:
    print('Image load failed!')
else:
    # 컬러 이미지의 모든 채널(B,G,R)에 100을 더함
    dst_color = cv2.add(src_color, (100, 100, 100, 0))

    # 원본 컬러 이미지와 밝기 조절 이미지를 각각 출력
    print('원본(컬러) 이미지')
    cv2_imshow(src_color)
    print('밝기 증가 후 이미지')
    cv2_imshow(dst_color)
```

cv2.add() 함수는 OpenCV에서 지원하는 이미지 덧셈 함수로, 컬러 이미지에 사용할 때 각 채널별로 값을 더해줍니다.

np.clip()과 .astype(np.uint8)를 통해 넘치거나(>255) 모자라는(<0) 값을 정상적인 이미지 범위로 맞춰줍니다.

Colab에서는 cv2.imshow() 대신 cv2_imshow()를 사용해야 이미지가 표시됩니다.
밝기값을 너무 크게 지정하면 일부 영역이 흰색(255)으로 포화될 수 있습니다.
